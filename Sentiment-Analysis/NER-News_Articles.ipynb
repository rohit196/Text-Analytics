```python
#Import Necessary Libraries
#%pip install lxml-html-clean newspaper3k
import spacy
from newspaper import Article
from spacy import displacy
```

```python
# Download and load the spaCy model for English Language
#python -m spacy download en_core_web_sm


nlp = spacy.load("en_core_web_sm")
```

```python
#Create Function to Extract Entities from Text
def extract_entities_from_url(url):
    #Download and prase the article
    article = Article(url)
    article.download()
    article.parse()

    #Extract the text from the article
    text = article.text

    # Apply NER using Spacy
    doc = nlp(text)

    # Extract and print the entities
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
    
    return entities
```

```python
#Example URL
url = "https://www.bbc.com/news/articles/cddyyyl4rm5o"
```

```python
#Extract Entities
entities = extract_entities_from_url(url)

#Display the entitites 
for entity in entities:
    print(f"Entity: {entity[0]}, Label: {entity[1]}")
```

```python
#Create function to extract and visualiza the entities from text
def extract_entities_with_visual(url):
    #Download and prase the article
    article = Article(url)
    article.download()
    article.parse()

    #Extract the text from the article
    text = article.text

    # Apply NER using Spacy
    doc = nlp(text)

    # Extract and print the entities
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    #Visualize the entities
    displacy.render(doc, style="ent", jupyter=True)

    return entities
```

```python
#Extract and Visualize Entities
entities = extract_entities_with_visual(url)
```
